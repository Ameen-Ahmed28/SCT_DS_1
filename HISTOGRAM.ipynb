```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Load the dataset
df = pd.read_csv("/content/drive/MyDrive/Skillcraft internship/Task 1 : Histogram   Bar Chart/StudentsPerformance (1).csv")
# Show dataset structure
print(df.head())  # Display the first few rows of the dataset
```

```python
# Plot histogram for a specific column
sns.histplot(df['reading score'], bins=20, kde=True, color='purple')  # Replace 'column_name' with the column to plot
plt.title("Reading score")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
# Using Matplotlib to create histograms for each numerical column
for column in df.select_dtypes(include=np.number):
    plt.hist(df[column], bins=20, color='blue', edgecolor='black', alpha=0.7)
    plt.title(f"Matplotlib Histogram for {column}")
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.grid(True)
    plt.show()
```
